```python
#Imports
from nltk.corpus import PlaintextCorpusReader
from matplotlib.colors import ListedColormap
from nltk.corpus import stopwords
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import pandas as pd
from os import path
import nltk
import os
```

```python
#seleção do arquivo
df = pd.read_excel('tweets_aqui.xlsx')
```

```python
#Separando palavras do arquivo em uma string
l = ' '.join(df.text.values)
lista = filter("https", l)
```

### Criando a nuvem 

"Stopwords" são as palavras como artigo, preposição etc que aparecem com recorrência e que você não quer levar em consideração na análise. Em geral, podemos dizer que são palavras "vazias", pois não carregam nenhum significados por elas próprias. Você pode consultar uma lista delas em qualquer idioma no exemplo abaixo.

```python
#stopwords.words('portuguese')
```

Aqui iremos criar a nuvem. Dentro da lista "stops" eu coloquei todas as stopwords da lingua portuguesa e adicionei algumas que eu também não gostaria de ver na minha nuvem (como gírias). Fique a vontade aqui para readaptar seu projeto.

```python
#Criando WordCloud
stops = stopwords.words('portuguese') + ['https']

mapa_cores = ListedColormap(['red']) #cor das palavras, pode ser mais de uma

wordcloud = WordCloud(background_color='white', #cor de fundo
                     colormap=mapa_cores,
                     stopwords=stops).generate(l)
```

Com a nuvem criada, vamos dar uma olhada antes de salvar!

```python
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
```

Salvando a nuvem como imagem no seu diretório:

```python
wordcloud.to_file("sua_cloud.png")
```
