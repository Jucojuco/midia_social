# Extração de dados da rede social Twitter

### Primeiros passos

```python
import tweepy
import pandas as pd

#suas credenciais Twitter aqui
consumer_key = ''
consumer_secret = ''
access_token = ''
access_token_secret = ''
```

```python
#acessando com suas credenciais
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth,wait_on_rate_limit=True)
```

### Definindo string de busca e extraindo 

String de busca:
Use "#Palavra" para buscar por palavras e #'s; use "@Palavra" para buscar um perfil; use " 'Pala vras' " para buscar expressões.
É recomendado não utilizar mais de 10 chaves de busca.
Sensível a letras maiúsculas e minúsculas

Laço de busca:
Defina máximo de tweets em "count"; a partir de qual data em "since" no formato do exemplo; adicione um filtro de idioma se preferir em "lang"

```python
data = []
strings = "'mapa do acolhimento' OR 'Mapa do Acolhimento' OR 'Mapa do acolhimento'"

for tweet in tweepy.Cursor(api.search, q=strings, count=20000,
                           since="2016-01-01", #).items():
                           lang='pt').items():
    data.append(tweet)
    
df_tweets = pd.DataFrame([q._json for q in data])
df_tweets.to_pickle('tweet.pkl')
```
